## 머신러닝이란?

>1. 데이터를 이해하는 알고리즘의 과학이자 애플리케이션
>2. **스스로 학습할 수 있는 머신러닝 알고리즘을 통해 해당 데이터를 지식으로 전환가능**



### 1. 데이터를 지식으로 바꾸는 지능적인 시스템 구축




>📌 
**머신러닝이 데이터에서 더 효율적으로 지식을 추출하여(사람이 수동적으로 작업하는 것보다)
예측 모델과 데이터 기반의 의사 결정 성능을 점진적으로 향상시킴**  

   
          - 일상 생활에서 많이 활용됨: 의료분야, 음성인식, 웹 검색 엔진,,,
      
      
### 2.  머신러닝의 세가지 종류  

  1️⃣ **지도 학습**  
  
        - 정답이 있는 데이터  
           
        - 정답과 출력을 비교하여 직접 피드벡 할 수 있음  
           
        - 출력을 예측  
           
           
           
  2️⃣ **비지도 학습**  
  
        - 정답(타깃)이 없음  
        
        - 정답이 없어서 직접 피드벡 할 수 없음  
        
        - 데이터의 숨겨진 구조/패턴을 학습  
           
           
  3️⃣ **강화 학습**  
  
        - 결정과정을 학습하는 것  
        
        - 타깃이 없음, 출력에 따른 보상을 주는 시스템  
        
        - 일련의 행동을 학습  
        
        
### 2-1. 지도 학습으로 미래 예측  



**🎯 주요 목적& 일련의 과정**   
    
  1. **레이블된 훈련 데이테에서 모델을 학습하여 본 적 없는 미래 데이터에 
      대해 예측을 만듦**  
  2. 레이블된 훈련 데이터가 머신 러닝 알고리즘에 전달되어 예측 모델을 훈련
      하고 그 다음 새로운 레이블되지 않은 데이터 입력에 대한 에측을 수행
      
(레이블: 머신러닝에서 특정 샘플에 할당된 클래스를 의미함)



- **분류**
> 1. 이진 분류  
>    **-두 개의 클래스 분류(양성 클래스: 내가 찾고자하는 클래스, 음성 클래스: 양성 클래스의 반대)**      
> 2. 다중 분류  
>    **-두 개보다 많은 클래스를 분류**  


- **회귀**   
> **-특성과 타깃이 주어졌을떄 출력 값을 예측하기 위해 두 변수 사이의 관계를 찾음**  

(선형 회귀: 데이터에서 학습한 직서의 기울기와 절편을 사용하여 새로운 데이터의 출력 값을 예측함)  



### 2-2. 비지도 학습으로 숨겨진 구조 발견

**🎯 주요 목적**   
    
**알려진 출력 값이나 보상 함수의 도움 없이 의미 있는 정보를 
추출하기 위해 데이터 구조/패턴을 탐색함.**  




- **군집: 서브그룹 찾기**
> **-사전 정보 없이 쌓여 있는 그룹 정보를 의미 있는 클러스터(서브그룹)로 
조직하는 탐색적 데이터 분석 기법**   


- **차원 축소: 데이터 압축**   
> **-잡음 데이터를 제거하기위해 관련 있는 데이터 정보를 대부분 유지하면서 
더 작은 차원을 가진 부분 공간으로 데이터를 압축함.** 



### 2-3. 강화 학습으로 반응형 문제 해결  

**🎯 주요 목적**   
    
**환경과 상호 작용하여 시스템 성능을 향상시키는 것이 목적.
(에이전트는 환경과 상호 작용하여 보상이 최대화되는 일련의 행동을 강화 학습으로 학습**  




 
### 3. 기본 용어와 표기법 소개

> 훈련 샘플: 데이터셋을 나타내는 테이블의 행  
  특성(x): 데이터 테이블이나 데이터 행렬의 열
  타깃(y): 결과, 출력
  손실 함수(비용함수): 해당 모델의 성능을 평가하는 지표(손실 함수 값이 낮을수록 좋은 성능으로 평가)
  

### 4. 기본 용어와 표기법 소개

*그림 첨부하기

-전처리: 데이터 형태 갖추기  

> 1. 선택된 특성들의 스케일 맞추기
> 2. 차원 축소
> 3. 훈련/테스트 데이터셋 나누기  

-예측 모델 훈련과 선택  

> 1. 가장 좋은 모델을 훈련하고 선택하기 위해 최소한 몇 가지 알고리즘을 비교해야함
> 2. 정확도: 정확히 분류된 비율
> 3. 교차 검증 기법: 모델의 일반화 성능을 예측하기 위해 훈련 데이터를 훈련/검증 데이터셋으로 나눔
> 4. 하이퍼파라미터 최적화: 좋은 성능을 내는 하이퍼파라미터(기계가 학습하는 것이 아닌 사람이 직접 선정해야하는 파라미터)를 찾아내야함  

-모델을 평가하고 본 적 없는 샘플로 예측  

> 훈련 데이터 셋에서 사용한 동일한 파라미터를 테스트 데이터셋은 물론 새로운 모든 샘플을 변환하는 데 사용해야함
